```python
graph = {
    'S' : ['A','B','C'],
    'A' : ['S','D'],
    'B' : ['D','S'],
    'C' : ['D','S'],
    'D' : ['A','B','C'],
}
 
def dfs(graph, node, visited):
    if node not in visited:
        visited.append(node)
        for k in graph[node]:
            dfs(graph,k, visited)
    return visited
 
visited = dfs(graph,'A', [])
print(visited)
```
