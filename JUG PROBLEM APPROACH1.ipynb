```python
def pour(jugM, jugN):
    A, B, fill = 3, 5, 4  
    print("%d\t%d" % (jugM,jugN))
    if jugN is fill:
        return
    elif jugN is B:
        pour(0, jugM)
    elif jugM != 0 and jugN is 0:
        pour(0, jugM)
    elif jugM is fill:
        pour(jugM, 0)
    elif jugM < A:
        pour(A, jugN)
    elif jugM < (B-jugN):
        pour(0, (jugM+jugN))
    else:
        pour(jugM-(B-jugN), (B-jugN)+jugN)
 
print("JUGM\tJUGN")
pour(0, 0)
```
